1. What is the difference between a function and a method in Python?
   - Function: A function is a block of reusable code that performs a specific task. It is defined using the def keyword. A function does not belong to any object or class unless placed inside one. its example is given below:

```python
def sum():
  a=int(input("Enter the first number: "))
  b=int(input("Enter the second number: "))
  print("Sum is: ",a+b)
sum()
```

- Method: A method is a function that is defined inside a class.It automatically takes the first parameter self (for instance methods).A method is called using an object of the class.It operates on the data (attributes) of that object.

```python
class Calculator:
  def add(self,a,b):
    return a+b
calculator=Calculator()
print(calculator.add(5,3))
```

- Key differences:
  - A function is not defined inside a class but a method is defined inside a class
  - first argument of a function is not anything special but that of a method should be self
  - a function is called using its name but a method is called using the object of the class its defined in.
  - a function operates on the data provided to it but a method operates on the class's data in which it is defined.
  - A function stands alone whereas a method belongs to the the class its defined in and operates on the objects data.

3. What are the different ways to define and call a function in Python?
   - there are many ways to define and call a function in python, some are defined as follows:
   - i. normal function
   - ii.function with default arguments
   - iii. Function with Positional Arguments
   - iv. Function with Keyword Arguments
   - v. Function with Variable-Length Arguments (*args)
   - vi. Function with Variable-Length Keyword Arguments (**kwargs)
   - vii. Function with Both *args and **kwargs
   - viii. Anonymous / Lambda Function
   - ix. Recursive Function
   - x. Nested Function
   All of these functions are explained with the help of examples as follows:
   

```python
# i. normal function
def sum(a,b):#function definition
  print("Sum with normal function:",a+b)
sum(10,5) #function call
print()

# ii.function with default arguments
def sum(a=10,b=20):#function definition with default arguments
  print("Sum with default arguments:",a+b)
sum() #function call
print()

# iii. Function with Positional Arguments
def student_details(name, age, course):#function definition with positional args
    print("Name:", name)
    print("Age:", age)
    print("Course:", course)
    print("Output with Positional Arguments")
student_details("Kulpreet", 20, "Python")#function call with ordered values
# here values should be in the accurate order
print()

# iv. Function with Keyword Arguments
def order(item, quantity, payment_mode):
    print("Item:", item)
    print("Quantity:", quantity)
    print("Payment Mode:", payment_mode)
    print("Output with Positional Arguments")
order(payment_mode="UPI", quantity=2, item="Pizza")
# as on the function call, the varibale names are given along with their values
# so that thier order donot matter
print()

# v. Function with Variable-Length Arguments (*args)
def sum(*args):#function definition with *args
    total = 0
    lis=[2,4,5.6,7,8.9]
    for num in lis:
        total += num
    print("Sum with Variable-Length Arguments:",total)
sum()    #function call
print()

# vi. Function with Variable-Length Keyword Arguments (**kwargs)
def details(**kwargs):
    print("Details with Variable-Length Keyword Arguments:")
    print(kwargs)
details(name="Kulpreet", age=20)
print()

# vii. Function with Both *args and **kwargs
def test(*args, **kwargs):
    print("Output with Both *args and **kwargs:")
    print(args, kwargs)
test(10, 20, name="Kulpreet")
print()

# viii. Anonymous / Lambda Function
square = lambda x: x * x
print("Using anonymous/lambda function")
square(5)
print()

# ix. Recursive Function
def fact(n):
    if n == 0:
        return 1
    return n * fact(n-1)
print("Using recursive function")
fact(5)
print()

# x. Nested Function
def outer():
    print("This is the outer function.")

    def inner():
        print("This is the inner function.")

    inner()   # calling inner function
print("Using nested function")
outer()
print()
```

6. Explain the concept of generators in Python and how they are defined.
   - A generator is a special type of iterator that generates values one at a time, instead of storing them in memory all at once. Generators are memory-efficient and are used for handling large data or infinite sequences.They use the yield keyword instead of return.
   - Key Features of Generators
     - Produce one value at a time
     - Use less memory
     - Maintain state between values
     - Automatically implement __iter__() and __next__()
     - Defined using a generator function or generator expression
     

```python
# Generators Are Defined: when it contains yield
def countdown(n):
  while n>0:
    yield n
    n-=1
gen= countdown(5)
print(next(gen))
```

7. What are the advantages of using generators over regular functions?
   - The first advantage of using generators over regular functions is that we recieve one value at a time, as it returns one value at a time.
   - second advantage is that it reduces the pressure on runtime to generate the full output at once, which could be very high in case where the number of iterations is huge and allows the runtime to generate result one at a time.
   - third advantage is that it uses memory efficiently, i.e., minimum at once, which could be exessively used when using regular functions. They Use Less Memory and do not store all values in memory. They produce one value at a time only when needed.
   - They Are Faster for Large Data. Since they don’t create the entire output at once, generators can be faster, especially with loops or large streams of data.
   - They Allow Infinite Sequences. Regular functions cannot return infinite lists. Generators can represent infinite streams.
   - They Support Lazy Evaluation. Generators compute values only when requested using next() or during iteration.This is called lazy evaluation, which reduces unnecessary computation.
   - They Make Pipelines Easy. Generators work well for data pipelines where each stage processes and yields data.
   

8. What is a lambda function in Python and when is it typically used?
   - A lambda function in Python is a small, anonymous function defined without using the def keyword. It is created using the keyword lambda and is typically used for short, simple operations.
   - Syntax: lambda arguments: expression

```python
# Example of lambda function
square = lambda x: x * x
print(square(9))
# Lambda function is used when:
# 1. You need a small function temporarily.Instead of defining a full function
# with def, you use lambda for quick tasks.
multiply = lambda a,b: a*b
print(multiply(45,78))
# 2. Used with functions like map(), filter(), reduce()
# Map function:
nums = [1, 2, 3, 4]
result = list(map(lambda x: x * 2, nums))
print(result)
# filter function:
even=list(filter(lambda x: x % 2 == 0,nums))
print(even)
# reduce function:
from functools import reduce
sum=reduce(lambda x,y:x+y,nums)
print(sum)
# 3. In places where defining a full function is unnecessary
def add(x): return x + 10
# Equivalent lambda:
add=lambda x: x + 10
print(add(50))
```

9. Explain the purpose and usage of the `map()` function in Python.
   - The map() function in Python is used to apply a function to every item in an iterable (like a list, tuple, or string) and return a new map object (which can be converted into a list, set, etc.).
   - Purpose of map():
   To process a sequence and transform each element using a function—without writing loops.
   - Syntax: map(function, iterables)
     - here function can be any in built or user defined function and iterables can be any list, string or any mutable data structure.
     - Following are some of the examples of using map function

```python
# using a defined function
def add(a):
  return a+10
list1=[1,2,3,4,5]
result=list(map(add,list1))
print(result)
```

```python
# using lambda function
list1=[1,2,3,4,5]
cube=list(map(lambda x:x**3,list1))
print(cube)
```

```python
# using multiple iterables
list1=[1,2,3,4,5]
list2=[6,7,8,9,10]
result=list(map(lambda x,y:x*y, list1,list2))
print(result)
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
    - These three functions are used for functional programming in Python.But each serves a different purpose:
    - 1. map() → Transform each element. its Purpose is: Applies a function to every item in an iterable and returns the modified result.
    - 2. filter() → Keep only items that meet a condition. its Purpose is: Returns only the elements that satisfy a condition (True/False).
    - 3. reduce() → Reduce the whole iterable to one value. its Purpose is:Applies a function repeatedly to combine all values into a single result.
    

```python
# map()- transforms values
nums = [1, 2, 3, 4]
result = list(map(lambda x: x * 2, nums))
print(result)
# filter()- removes unwanted values
nums = [1, 2, 3, 4, 5]
result = list(filter(lambda x: x % 2 == 0, nums))
print(result)
# reduce()- collapses a list into one result
from functools import reduce
nums = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, nums)
print(result)
```

```python
#  5. Create an iterator class in Python that generates the Fibonacci sequence
# up to a specified number of terms.
class fibonacci_iterator:
  def __init__(self,n):
    self.n=n
    self.count=0
    self.a=0
    self.b=1

  def __iter__(self):
    return self

  def __next__(self):
    if self.count>self.n:
      raise StopIteration

    if(self.count==0):
      self.count+=1
      return self.a

    elif(self.count==1):
      self.count+=1
      return self.b


    next_num=self.a+self.b
    self.a=self.b
    self.b=next_num

    self.count+=1
    return next_num

for num in fibonacci_iterator(15):
  print(num,end="  ")
```
