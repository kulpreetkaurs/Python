2. How does broadcasting work in NumPy?
   - broadcasting in Numpy:Broadcasting in NumPy refers to the ability to perform arithmetic operations on arrays of different shapes. This feature allows NumPy to treat arrays with different dimensions during arithmetic operations, ensuring they have compatible shapes without making unnecessary copies of data.
   - Broadcasting works by stretching the smaller array across the larger array so that they have compatible shapes. This process is efficient because it avoids making needless copies of data and allows operations to be vectorized, which means they are executed in C rather than Python.
   - Example of Broadcasting with Scalars: When an array and a scalar value are combined in an operation, the scalar is stretched to match the shape of the array:


```python
import numpy as np
a = np.array([1.0, 2.0, 3.0])
b = 2.0
result = a * b
print(result)
```

4. Explain the use of the groupby() method in Pandas?
   - the groupby() method in Pandas is used to split data into groups based on one or more keys (columns or index levels), apply a function to each group, and then combine the results into a new DataFrame or Series.It follows the Split-Apply-Combine strategy:
   - Syntax: DataFrame.groupby(by=None, axis=0, level=None, as_index=True, sort=True, group_keys=True)
   - Key Parameters:
    - by: Column name(s), index level(s), function, dictionary, or Series to group by.
      - axis: 0 for rows (default), 1 for columns.
      - as_index: If True, group keys become the index in the result.
      - sort: Sort group keys (default True).
      - group_keys: If True, adds group labels to the index.
   - How It Works:
   Split - Divide the DataFrame into groups based on the by parameter.
   Apply - Apply aggregation (sum, mean, count, etc.), transformation, or filtering to each group.
   Combine - Merge the results back into a DataFrame or Series.


```python
#Example
import pandas as pd
data = {
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR', 'Finance'],
    'Employee': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Salary': [60000, 50000, 65000, 70000, 52000, 72000]
}
df = pd.DataFrame(data)
avg_salary = df.groupby('Department')['Salary'].mean()
print(avg_salary)
```

6. What are the differences between NumPy arrays and Python lists?
   - Performance
     - NumPy Arrays: NumPy arrays are highly optimized for numerical computations. They use contiguous memory allocation and are implemented in C, which allows for faster execution of mathematical operations. For example, element-wise operations and matrix manipulations are significantly faster compared to Python lists.
     - Python Lists: Python lists are slower for numerical computations due to their dynamic and heterogeneous nature. Each element in a list carries additional metadata, and operations are interpreted at runtime, leading to higher overhead.
    - Functionality
      - NumPy Arrays: NumPy arrays are homogeneous, meaning all elements must be of the same data type. This uniformity enables efficient memory usage and specialized operations like broadcasting, slicing, and advanced indexing. They also support a wide range of mathematical functions and linear algebra operations.
      - Python Lists: Python lists are heterogeneous, allowing elements of different data types. While this makes them versatile for general-purpose use, they lack built-in support for numerical operations. Element-wise operations require explicit loops or external libraries, making them less convenient for scientific computing.
    - Memory Efficiency
      - NumPy Arrays: NumPy arrays consume less memory as they store elements in contiguous memory blocks and avoid storing type information for each element. For instance, a NumPy array of 1000 integers uses approximately 8000 bytes, compared to 48000 bytes for a Python list of the same size.
      - Python Lists: Lists are memory-inefficient because they store additional metadata for each element, such as type and reference count. This overhead increases significantly with larger datasets.
    - Ease of Use
      - NumPy Arrays: NumPy arrays are more convenient for numerical tasks due to their extensive library of functions and ability to perform operations without explicit loops. For example, adding a scalar to all elements of an array is straightforward.
      - Python Lists: Lists require manual iteration for operations like adding a scalar to each element, making them less user-friendly for numerical tasks.
    - Conclusion NumPy arrays are ideal for performance-critical and numerical applications, while Python lists are better suited for general-purpose programming where flexibility is required.

8. What does the term “vectorized operation” mean in NumPy?
   - In NumPy, a vectorized operation means performing a mathematical or logical operation on entire arrays at once, without writing explicit Python loops (for or while).Instead of iterating element-by-element in Python (which is slow), NumPy uses highly optimized C and Fortran code under the hood to apply the operation to all elements in parallel, making it much faster and more concise.
   - Benefits of Vectorization in NumPy
     - Speed - Operations run much faster because they use low-level optimized code.
     - Cleaner Code - No need for verbose loops.
     - Memory Efficiency - Works directly on contiguous memory blocks.
     - Parallelization - Many operations can be executed in parallel internally.

```python
#Example without vectorized operation
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
result = np.zeros_like(arr)
for i in range(len(arr)):
    result[i] = arr[i] ** 2
print(result)  # [ 1  4  9 16 25 ]
```

```python
#Example with vectorized operation
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
result = arr ** 2
print(result)  # [ 1  4  9 16 25 ]
```

    - When working with pandas, missing values (NaN, None, pd.NA) can disrupt calculations and cause unexpected results. Instead of halting processing, pandas provides robust tools to detect, remove, or fill these gaps so analysis can proceed smoothly.

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
   'A': [1, 2, np.nan, 4],
   'B': [np.nan, 2, 3, np.nan]
})
# Detect missing values
print(df.isna())
# Fill missing values with column mean
df_filled = df.fillna(df.mean(numeric_only=True))
print(df_filled)
```

15. How does NumPy handle multidimensional arrays?
    - NumPy handles multidimensional arrays using its core data structure called the ndarray (N-dimensional array). This structure is highly optimized for numerical computations and can store elements of the same data type in a contiguous block of memory, enabling fast indexing, slicing, and vectorized operations.
    - 1. Creation: You can create multidimensional arrays using:

```python
import numpy as np
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
arr_3d = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
# ndim → Number of dimensions
# shape → Tuple showing size along each dimension
# size → Total number of elements
print(arr_2d.ndim)
print(arr_2d.shape)
print(arr_2d.size)
```

```python
# 2. Indexing
# NumPy uses row-major order (C-style) by default, meaning the last axis
# changes fastest in memory.
print(arr_2d[0, 1])
print(arr_3d[1, 0, 1])
```

```python
# 3. Slicing
# You can extract subarrays without copying data (views):
sub = arr_2d[:, 1:]
print(sub)
# Changes in sub will reflect in arr_2d unless you explicitly copy.
```

```python
# 4. Broadcasting
# NumPy can perform operations between arrays of different shapes by broadcasting:
a = np.array([[1], [2], [3]])
b = np.array([10, 20, 30])
print(a + b)
```

```python
# 5. Reshaping
# You can change the shape without changing the data:
reshaped = arr_2d.reshape(3, 2)
print(reshaped)
# 6. Memory Efficiency
# NumPy stores data in contiguous memory blocks.
# Indexing and slicing return views (no data copy) for efficiency.
# Uses strides to map multidimensional indices to memory addresses.
```

17. Explain the difference between apply() and map() in Pandas?
    - In Pandas, both apply() and map() are used to apply functions to data, but they differ in scope, usage, and flexibility.
    - 1. map(): Scope: Works only on a Pandas Series (1D data).Purpose: Apply a function, dictionary mapping, or Series mapping element-wise.Usage: Simple transformations or value mapping.
    - Key Points:
      - Element-wise operation.
      - Can take a function, dict, or Series as an argument.
      - Cannot be directly applied to a DataFrame.
    - 2. apply(): Scope: Works on both Series and DataFrame. Purpose: Apply a function along an axis (rows or columns) for DataFrames, or element-wise for Series.Usage: More flexible, supports complex operations.
    - Key Points:
      - Works on Series (element-wise) and DataFrame (row/column-wise).
      - Accepts only a function (no dict or Series mapping).
      - More powerful for aggregations and transformations.

```python
#Example of map function
import pandas as pd
s = pd.Series([1, 2, 3, 4])
# Using a function
print(s.map(lambda x: x**2))
# Using a dictionary mapping
print(s.map({1: 'A', 2: 'B'}))
# Using a Series mapping
print(s.map(pd.Series(['X', 'Y', 'Z'], index=[1, 2, 3])))
```

```python
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})
# Apply to a Series (similar to map)
print(df['A'].apply(lambda x: x**2))
# Apply to a DataFrame column-wise
print(df.apply(lambda col: col.max(), axis=0))
# Apply to a DataFrame row-wise
print(df.apply(lambda row: row.sum(), axis=1))
```

20. What is the role of a pivot table in Pandas?
    - A pivot table is a data analysis tool that allows you to take columns of raw data from a Pandas DataFrame, summarize them, and analyze the summary data to reveal insights. It enables you to perform common aggregate statistical calculations such as sums, counts, averages, and more, which can help uncover trends and patterns that may not be immediately apparent in the raw data.
    - Key Functions of Pivot Tables in Pandas
      - Data Aggregation: Pivot tables allow you to aggregate data based on specified columns. You can calculate various statistics (e.g., sum, mean, count) for different groups within your data.
      - Data Reshaping: They enable you to reshape your data, making it easier to analyze. For example, you can transform long-format data into a wide format, where unique values from one column become new columns in the pivot table.
      - Multi-dimensional Analysis: Pivot tables can handle multiple dimensions, allowing you to group data by multiple categories. This is useful for comparing different segments of your data side by side.
      - Custom Aggregation Functions: You can define custom aggregation functions to apply to your data, providing flexibility in how you summarize your information.
    - How to Create a Pivot Table in Pandas
      - To create a pivot table in Pandas, you can use the pivot_table() function. Here’s a basic syntax overview:
      - pivot_table = pd.pivot_table(data, values='column_to_aggregate', index='column_to_group_by', columns='column_to_create_subcategories', aggfunc='mean')
      - data: The DataFrame you want to analyze.
      - values: The column(s) to aggregate.
      - index: The column(s) to group by (rows).

21. Why is NumPy's array slicing faster than Python's list slicing?
    - NumPy's array slicing is significantly faster than Python's list slicing due to differences in memory layout, data type handling, and the underlying implementation of operations.
    - Memory Layout: NumPy arrays store data in contiguous memory blocks, which allows for efficient access and manipulation. This is known as locality of reference. In contrast, Python lists are collections of pointers to objects, which are stored in non-contiguous memory locations. Accessing elements in a list requires dereferencing these pointers, adding overhead . For example:A NumPy array directly accesses the data in memory.A Python list must traverse pointers to access the actual data.This contiguous memory layout in NumPy enables faster slicing and operations, as the data can be processed in bulk without additional pointer lookups.
    - Homogeneous Data Types: NumPy arrays are homogeneous, meaning all elements share the same data type. This allows NumPy to use fixed-size memory blocks and optimized low-level operations. Python lists, on the other hand, are heterogeneous, meaning they can store elements of different types. This flexibility comes at the cost of performance, as additional type checks and memory management are required during slicing.
    - Underlying Implementation: NumPy is implemented in C, and its operations are highly optimized for performance. When slicing a NumPy array, the operation is performed at the C level, bypassing Python's runtime overhead. Python lists, however, rely on Python's interpreted runtime, which is inherently slower.
    - Vectorized Operations: NumPy supports vectorized operations, allowing entire slices to be processed in parallel using SIMD (Single Instruction, Multiple Data) techniques. Python lists do not support such optimizations, as operations are performed element by element in Python loops.
    - Practical Implications: NumPy's slicing is ideal for numerical and scientific computations where large datasets are involved. It is particularly advantageous in tasks like matrix operations, image processing, and machine learning, where performance is critical.
    - Limitations:While NumPy is faster for slicing and numerical operations, it is less flexible than Python lists for handling mixed data types or dynamic resizing. For non-numerical tasks, Python lists may still be more appropriate.
